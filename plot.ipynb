## Importing

```python
import matplotlib.pyplot as plt
import matplotlib.colors as mpltcolor
```

## Line Graph

```python
#Title
plt.title('My first Graph')

#Axis
plt.xlabel('X axis')
plt.ylabel('Y axis')

#Values
x = [1, 2]
y = [2, 3]

plt.plot(x,y)
plt.show()
```

## Bar graph

```python
#Title
plt.title('Bar Graph')

#Values
x = [1, 2, 3, 4, 5]
y = [2, 3, 7, 1, 0]

#Axis
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.bar(x, y)
plt.show()
```

### Matching two bar graphs and adding legend

```python
#Title
plt.title('Matching two bar graphs')

#Values
x1 = [1, 3, 5, 7, 9]
y1 = [2, 3, 7, 2, 1]

x2 = [2, 4, 6, 8, 10]
y2 = [3, 4, 5, 1, 2]

#Axis
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.bar(x1, y1, label='Group one')
plt.bar(x2, y2, label='Group two')
plt.legend() #Add legend 
plt.show()
```

## Scatterplot

```python
#Title
plt.title('Scatterplot')

#Values
x = [1, 2, 3, 4, 5]
y = [2, 3, 7, 1, 0]

plt.scatter(x, y, label='My dots', color='g', marker='.', s=100) #Create the scatterplot
plt.plot(x, y, color='r', linestyle='--') #Create a line graph that connects all dots
plt.legend()
plt.show()
#plt.savefig('Scatterplot.pdf', dpi=300) #Save as pdf
```

## 1980-2016 brazilian population

```python
#Data getted from DataSUS
data = open('brazilian_population.csv').readlines()

#Graph values
years = []
popul = []

#Using for loop to get the values
for i in range(len(data)):
    if i != 0:
        row = data[i].split(';')
        years.append(int(row[0]))
        popul.append(int(row[1]))

plt.title('1980-2016 Brazilian Population Growth')
plt.bar(years, popul, color='#00FF49')
plt.plot(years, popul, color='b', linestyle=':')
plt.xlabel('Year')
plt.ylabel('Population (Million)')
plt.show()
```

## Exercise: Comparing human and bacteria genome

### Bacteria

```python
bacteriaEntry = open('bacteria.fasta').read() #Get Values from .fasta file

bacteriaEntry = bacteriaEntry.replace('\n', '') #Removing /n 

counter = {} #Create a counter dict 

for c in ['A', 'T', 'C', 'G']:
    for j in ['A', 'T', 'C', 'G']:
        counter[c+j] = 0 #concat two strings to get a genome

for k in range(len(bacteriaEntry)-1):
    counter[bacteriaEntry[k]+bacteriaEntry[k+1]] += 1 
```

### Bacteria HTML Output 

```python
bacteriaOutput = open('bacteria.html', 'w') #Outputing de data in a .html file
i = 1
for k in counter:
    alpha = counter[k]/max(counter.values()) #Calculate a value betwen 0-1 to alpha, Objetive: the higher value is more visible
    rgbaColor = [0, 0, 255, alpha] #Creating rgba color
    bacteriaOutput.write(f'<div style="width:100px; border:1px solid #111; height:100px; color:#ffffff;float:left; background-color:rgba{tuple(rgbaColor)}">{k}</div>') #Write a html document 

    if i%4 == 0: #Verify if have 4 blocks 
        bacteriaOutput.write("<div style='clear:both'></div>") #Break line when has 4 blocks created
    i += 1

bacteriaOutput.close()
```
